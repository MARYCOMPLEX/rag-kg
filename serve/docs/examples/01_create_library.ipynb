# Example 01 — Create your first Library

This notebook walks through the smallest possible interaction with the API:
creating a Library and listing what you have.

Pre-reqs:
1. Backend running: `uv run uvicorn apps.api.main:app --reload`
2. (Optional) `API_KEY` set in `.env` — same value here as `Authorization: Bearer ...`

```python
import httpx

BASE = 'http://localhost:8000'
API_KEY = ''  # optional — match Settings.api_key
headers = {'Authorization': f'Bearer {API_KEY}'} if API_KEY else {}
client = httpx.Client(base_url=BASE, headers=headers, timeout=30)
```

## Create a Library

```python
res = client.post(
    '/v1/libraries',
    json={
        'library_id': 'demo',
        'name': 'Demo Library',
        'description': 'Created from the example notebook',
    },
)
print(res.status_code, res.json())
```

## List Libraries

```python
res = client.get('/v1/libraries')
for lib in res.json():
    print(f"{lib['library_id']:20}  {lib['name']:30}  {lib.get('description') or ''}")
```
